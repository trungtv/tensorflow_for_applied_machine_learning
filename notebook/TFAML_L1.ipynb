## Initiate a session: Tf.session()

```python
import tensorflow as tf #import tensorflow library
# Build a graph.
a = 5 #tf.constant(5.0) 
b = tf.constant(6.0) 
c=a*b
# Launch the graph in a session.
with tf.Session() as sess: 
    # Evaluate the tensor `c`.
    print(sess.run(c))
    
# sess will be closed automatically after "with block"
```

## Feeding the execution graph
A graph can be parameterized to accept external inputs  
A placeholder is a promise to provide a value later, like a function argument

```python
x = tf.placeholder(tf.float32) 
y = tf.placeholder(tf.float32) 
z=x+y
with tf.Session() as sess:
    print(sess.run(z, feed_dict={x: 3, y: 4.5}))
    print(sess.run(z, feed_dict={x: [1, 3], y: [2, 4]}))
    print(sess.run(z, feed_dict={x: 30, y: 4.5}))
```

## tf.InteractiveSession  
an InteractiveSession installs itself as the default session on construction. The methods tf.Tensor.eval and tf.Operation.runwill use that session to run ops.


```python
sess = tf.InteractiveSession()
a = tf.constant(5.0)
b = tf.constant(6.0)
c = a * b
# We can just use 'c.eval()' without passing 'sess'
print(c.eval())    
print(sess.run(c))
sess.close()
```

## tf.cond(pred, fn1, fn2, name=None)
Return either fn1() or fn2() based on the boolean predicate pred.
fn1 and fn2 both return lists of output tensors. fn1 and fn2 must have the same non-zero number and type of outputs.

```python
import tensorflow as tf #import tensorflow library
a = 10
b = 2
x = tf.Variable(5)
y = tf.Variable(30)
z = tf.multiply(a, b)
result = tf.cond(x < y, lambda: tf.add(x, z), lambda: tf.square(y))
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    print(result.eval())
```

```python
# Tf.Variable
number = tf.Variable(2)
multiplier = tf.Variable(1)
multiplier = multiplier.assign_add(1)
init = tf.global_variables_initializer()
result = tf.multiply(number,multiplier)
with tf.Session() as sess:
    sess.run(init)
    for i in range(10):
        x, y, z = sess.run([number, multiplier,result])
        print("Result number: {} * multiplier: {} = {}".format(x, y, z))
```

```python
# Tf.constant
number = tf.constant(2)
multiplier = tf.constant(1)
multiplier = tf.add(multiplier, 1)
result = tf.multiply(number,multiplier)
with tf.Session() as sess:
    sess.run(init)
    for i in range(10):
        print("Result number * multiplier = ",sess.run(result))
```
